# <span style="color:#336699">Introdução</span>
<hr style="border:1px solid #0077b9;">

A biblioteca ``ipyleaflet`` permite a criação de mapas interativos diretamente no Jupyter Notebook. Ela é uma ponte entre o Python e a popular biblioteca JavaScript Leaflet.js, sendo excelente para visualização de dados geoespaciais, imagens de satélite e geometrias vetoriais.

Para aprender a criar visualizações com o ``ipyleaflet``, é importante seguir um fluxo de trabalho básico composto por 6 etapas principais:

1 - Criar o Mapa base

2 - Criar Camadas (Layers) e, opcionalmente, anexar Popups

3 - Adicionar as Camadas ao Mapa

4 - Configurar Propriedades visuais

5 - Lidar com Interações do usuário

6 - Gerenciamento de Camadas (remoção/limpeza)

# <span style="color:#336699">Importar bibliotecas</span>
<hr style="border:1px solid #0077b9;">

Por convenção, importamos as classes principais do ``ipyleaflet`` e também módulos do ``ipywidgets`` para nos auxiliar com as interfaces:

```python
from ipyleaflet import Map, Marker, Polygon, GeoJSON, basemaps
from ipywidgets import HTML
import json
```

# <span style="color:#336699">1 - Criando o Mapa</span>
<hr style="border:1px solid #0077b9;">

A primeira etapa de qualquer visualização no ``ipyleaflet`` é instanciar o objeto ``Map``.

Para isso é preciso definir, no mínimo, as coordenadas centrais (``center``), o nível de zoom (``zoom``) e o mapa de fundo (``basemap``).

Vamos centralizar o mapa no INPE de São José dos Campos:

```python
# Definindo as coordenadas (Latitude, Longitude)
coords_sjc = (-23.2085, -45.8696)

# Instanciando o mapa
m = Map(
    center=coords_sjc,
    zoom=13,
    basemap=basemaps.Esri.WorldTopoMap
)

# Exibindo o mapa na célula (basta chamar a variável)
m
```

Façam teste substituindo o ``basemap`` por ``basemaps.Esri.WorldImagery`` ou ``basemaps.CartoDB.DarkMatter``.

# <span style="color:#336699">2 - Criando Camadas</span>
<hr style="border:1px solid #0077b9;">

Com o mapa base criado, o próximo passo é criar as camadas que sobrepõem os dados.

Podemos criar Pontos (``Marker``), Polígonos (``Polygon``), ou carregar dados estruturados como ``GeoJSON``.

<b>Criando um Ponto</b>

```python
marcador_inpe = Marker(
    location=coords_sjc,
    draggable=False,
    title="Sede do INPE"
)
```

<b>Criando um Polígono</b>

Vamos criar uma geometria poligonal simples para representar uma área  ou delimitação de interesse:

```python
# Lista de coordenadas formando um polígono fechado
area_estudo = [
    (-23.2000, -45.8800),
    (-23.2000, -45.8600),
    (-23.2200, -45.8600),
    (-23.2200, -45.8800)
]

poligono_inpe = Polygon(
    locations=area_estudo,
    color="blue",       # Cor da borda
    fill_color="blue"   # Cor de preenchimento
)
```

<b>Criando e Anexando Popups</b>

ma funcionalidade muito útil é a exibição de informações quando o usuário clica em um elemento do mapa.

No ``ipyleaflet``, isso é feito criando um widget ``HTML`` e anexando-o à propriedade popup da camada.

```python
# Criando o conteúdo em HTML
conteudo_popup = HTML()
conteudo_popup.value = """
    <b>INPE - Sede principal</b><br>
    <i>São José dos Campos, SP</i><br>
    Pesquisa e Desenvolvimento em Observação da Terra.
"""

# Associando o widget HTML ao popup do marcador que criamos anteriormente
marcador_inpe.popup = conteudo_popup
```

# <span style="color:#336699">3 - Adicionando Camadas ao Mapa</span>
<hr style="border:1px solid #0077b9;">

Até agora, o ponto e o polígono existem apenas na memória do Python. Eles não aparecerão no mapa até que você os adicione.

Para isso vamos utilizar o método ``add()`` (ou ``add_layer()``) ou o operador ``+=`` para inserir as camadas no mapa ``m``.

```python
# Adicionando o marcador com popup
m.add(marcador_inpe)

# Adicionando a área do polígono
m.add(poligono_inpe)

# Verificando novamente o mapa atualizado
m
```

# <span style="color:#336699">4 - Configurando Propriedades</span>
<hr style="border:1px solid #0077b9;">

Uma vantagem de rodar o ``ipyleaflet`` no Jupyter é que as propriedades do mapa e das camadas estão ligadas.

Isso significa que você pode alterar propriedades como cor e opacidade depois da camada ter sido criada e adicionada, e o mapa será atualizado instantaneamente na tela, sem precisar ser recarregado!

```python
# Alterando as propriedades do polígono já renderizado
poligono_inpe.color = "red"             # Muda a borda para vermelho
poligono_inpe.fill_color = "orange"     # Muda o preenchimento para laranja
poligono_inpe.fill_opacity = 0.5        # Deixa a área semi-transparente
```

# <span style="color:#336699">5 - Lidando com Interações</span>
<hr style="border:1px solid #0077b9;">

Mapas interativos reagem às ações do usuário. O método ``on_interaction()`` permite que você registre uma função de retorno (``callback``) que será executada sempre que ocorrer um evento, como um clique ou movimento do mouse no mapa.

Vamos criar uma interação que captura a coordenada exata onde o usuário clicou no mapa e imprime (``print``) essa informação:

```python
# 1. Definimos a função de callback
def ao_clicar_no_mapa(**kwargs):
    # kwargs captura vários eventos do mapa.
    # O evento 'type' nos diz qual ação ocorreu.
    if kwargs.get('type') == 'click':
        latlon = kwargs.get('coordinates')
        print(f"Você clicou nas coordenadas: Latitude {latlon[0]:.4f}, Longitude {latlon[1]:.4f}")

# 2. Registramos a interação no objeto do mapa
m.on_interaction(ao_clicar_no_mapa)

# Observação: Clique no mapa exibido nas células anteriores para ver o texto aparecer abaixo da célula onde o mapa está renderizado.
```

# <span style="color:#336699">6 - Gerenciamento de Camadas</span>
<hr style="border:1px solid #0077b9;">

Com o passar da análise, seu mapa pode ficar poluído com muitos dados.

A etapa final do nosso aprendizado de ipython é o gerenciamento dessas camadas, o que inclui a remoção de elementos específicos ou a limpeza total do mapa.

<b>Removendo uma camada específica</b>

Se quisermos remover apenas o polígono de área de estudo, mas manter o marcador.

```python
# Remove especificamente a camada poligonal
m.remove(poligono_inpe)
```

<b>Limpando todas as camadas</b>

```python
# Remove todas as geometrias, marcadores e dados inseridos
m.clear_layers()
```

Observe como o ``ipyleaflet`` é modular. Ao separar a criação de camadas da sua adição ao mapa e posterior configuração, a gente ganha mais controle para desenvolver dashboards e ferramentas de análise geoespacial.

# <span style="color:#336699">Trabalhando com Dados Raster (Mosaico de Imagens de Satélite)</span>
<hr style="border:1px solid #0077b9;">

Além de marcadores e polígonos vetoriais, o `ipyleaflet` permite a visualização de dados matriciais (Raster) de forma muito eficiente.

Neste exercício, vamos visualizar um mosaico completo do Brasil construído com imagens do satélite **Sentinel-2**. Esses dados são disponibilizados pelo **Brazil Data Cube (BDC)**.

Como um mosaico do Brasil inteiro é um arquivo gigantesco, nós **não** vamos baixar a imagem. Em vez disso, faremos uso do serviço **WMS (Web Map Service)**.

**Quais as vantagens do WMS?**
* **Muito mais rápido e leve:** O servidor do INPE faz todo o trabalho pesado. O seu computador não sofre para carregar os dados.
* **Inteligente:** O mapa baixa automaticamente apenas os "quadradinhos" da imagem (tiles) que você está olhando na tela naquele exato momento e zoom.
* **Código simples:** Não precisamos empilhar matrizes ou calcular recortes, basta informar a URL do servidor e o nome da camada!

```python
from ipyleaflet import Map, basemaps, WMSLayer

# 1. Cria o mapa centralizado no Brasil
m = Map(center=(-15.0, -50.0), zoom=4, basemap=basemaps.Esri.WorldTopoMap)

# 2. Configura a camada WMS (apenas URL base e camada)
wms_mosaico = WMSLayer(
    url="https://data.inpe.br/bdc/geoserver/mosaics/wms",
    layers="mosaics:mosaic-s2-brazil-3m",
    format="image/png",
    transparent=True
)

# 3. Adiciona a camada ao mapa e exibe
m.add(wms_mosaico)
m
```

# <span style="color:#336699">Exibindo Múltiplos Mapas de Uso e Cobertura da Terra com Controlador</span>
<hr style="border:1px solid #0077b9;">

```python
from ipyleaflet import Map, basemaps, WMSLayer, LayersControl

# 1. Mapa base
m = Map(center=(-3.5, -62.0), zoom=6, basemap=basemaps.Esri.WorldTopoMap)

# 2. TerraClass Amazônia v3 (2020)
terraclass = WMSLayer(
    url="https://data.inpe.br/bdc/geoserver/terraclass/ows/terraclass/wms",
    layers="terraclass-amazonia-v3",
    name="TerraClass Amazônia 2020",
    format="image/png",
    transparent=True,
    time="2020"
)

# 3. MapBiomas Brasil v10 (2020)
mapbiomas = WMSLayer(
    url="https://data.inpe.br/bdc/geoserver/mapbiomas/ows/mapbiomas/wms",
    layers="mapbiomas-brasil-v10",
    name="MapBiomas Brasil 2020",
    format="image/png",
    transparent=True,
    time="2020"
)

# 4. IBGE Cobertura e Uso da Terra (2020)
ibge = WMSLayer(
    url="https://data.inpe.br/bdc/geoserver/ibge/ows/ibge/wms",
    layers="cobertura_uso_terra",
    name="IBGE Uso da Terra 2020",
    format="image/png",
    transparent=True,
    PropertyName="classe_id_uso2020"
)

# 5. Adicionando as camadas ao mapa
m.add(terraclass)
m.add(mapbiomas)
m.add(ibge)

# 6. Adicionando o controlador de camadas
control = LayersControl(position='topright')
m.add_control(control)

# Exibindo o mapa
m
```

# <span style="color:#336699">Verificando sobreposição Empreendimentos x Terra Indígena</span>
<hr style="border:1px solid #0077b9;">

Com as recentes atualizações do Conselho Monetário Nacional (como as Resoluções CMN Nº 5.193/2024, 5.267/2025 e 5.268/2025), a verificação de financiamentos tornou-se mais rigorosa. Hoje, exige-se a identificação da propriedade via Cadastro Ambiental Rural (CAR) e o monitoramento obrigatório por sensoriamento remoto para áreas superiores a 300 hectares, garantindo que os recursos financiados respeitem a conformidade socioambiental.

###  Terras Indígenas

Terra Indígena, conforme a Constituição Federal de 1988, é um território demarcado e protegido para a posse permanente e o usufruto exclusivo dos povos indígenas. Essas terras são reconhecidas como patrimônio da União e são destinadas à preservação de sua cultura, tradições, recursos naturais e formas de organização social, além de assegurar a reprodução física e cultural dessas comunidades. A demarcação das terras indígenas é um direito constitucional e visa garantir a autodeterminação, a autonomia e a proteção dos direitos dos povos indígenas, bem como sua participação ativa na gestão e preservação desses territórios.

De acordo com a Resolução CMN Nº 5.193/2024:

> 7 - Não será concedido crédito rural para empreendimento situado em imóvel rural total ou parcialmente inserido em
terras ocupadas por indígenas, observado que: (Res CMN 5.193 art 1º)
a) as terras ocupadas por indígenas devem constar como homologadas, regularizadas ou definidas como Reserva
Indígena no Sistema Indigenista de Informações da Fundação Nacional dos Povos Indígenas (Funai); e
b) o disposto no caput não se aplica aos casos em que o proponente pertença aos povos ou às comunidades
indígenas ocupantes ou habitantes da terra indígena na qual se situa o empreendimento.

**Exemplo com dados reais**

```python
r = requests.get("https://raw.githubusercontent.com/GSansigolo/shapefiles/refs/heads/main/mini_tis_ma.zip")

zipfile.ZipFile(io.BytesIO(r.content)).extractall("./mini_tis_ma.zip")
file_path = "./mini_tis_ma.zip/mini_tis_ma.shp"

mini_tis = gpd.read_file(file_path)
```

```python
r = requests.get("https://raw.githubusercontent.com/GSansigolo/shapefiles/refs/heads/main/poligono_3.zip")

zipfile.ZipFile(io.BytesIO(r.content)).extractall("./poligono_3.zip")
file_path = "./poligono_3.zip/poligono_3.shp"

poligono_3 = gpd.read_file(file_path)
```

```python
from ipyleaflet import Map, GeoData, LegendControl, basemaps

# Pegar os limites extremos do empreendimento: [minx, miny, maxx, maxy]
limites = poligono_3.total_bounds

# Calcular o centro do mapa baseado na bounding box (evita o aviso do geopandas)
lon_centro = (limites[0] + limites[2]) / 2
lat_centro = (limites[1] + limites[3]) / 2

# Cria o mapa base interativo usando o CartoDB Positron
m = Map(
    basemap=basemaps.CartoDB.Positron,
    center=(lat_centro, lon_centro),
    zoom=7
)

# Criar a camada da Terra Indígena (mini_tis)
camada_tis = GeoData(
    geo_dataframe=mini_tis,
    style={
        'color': '#d62728',
        'weight': 2,
        'fillColor': '#d62728',
        'fillOpacity': 0.3
    },
    name='Terra Indígena'
)

# Criar a camada do Empreendimento (poligono_3)
camada_empreendimento = GeoData(
    geo_dataframe=poligono_3,
    style={
        'color': '#1f77b4',
        'weight': 2,
        'fillColor': '#1f77b4',
        'fillOpacity': 0.3
    },
    name='Empreendimento'
)

# Adiciona as camadas ao mapa
m.add_layer(camada_tis)
m.add_layer(camada_empreendimento)

# Criar e adicionar a Legenda
legenda = LegendControl(
    {
        "Empreendimento": "#1f77b4",
        "Terra Indígena": "#d62728"
    },
    title="Legenda",
    position="bottomright"
)
m.add_control(legenda)

# Exibe o mapa
m
```

```python
poligono_3 = poligono_3.to_crs(mini_tis.crs)
geom_empreendimento = poligono_3.geometry.iloc[0]

mini_tis_plano = mini_tis.to_crs(epsg=6933)
geom_empreendimento_plana = poligono_3.to_crs(epsg=6933).geometry.iloc[0]

sobreposicoes = mini_tis.intersects(geom_empreendimento)
areas = mini_tis_plano.intersection(geom_empreendimento_plana).area

area_minima = 1.0

for i, sobrepoe in sobreposicoes.items():
    if sobrepoe and areas[i] > area_minima:
        print(f"O empreendimento sobrepõe a área da Terra Indígena {i}! Área da sobreposição: {areas[i]:.2f} m².")
```
